```python
import onnx
import caffe2.python.onnx.backend
import numpy as np
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original', data_home="dataset")


```

```python
# Load the ONNX model
model = onnx.load('mnist.onnx')
# Run the ONNX model with Caffe2
for d in mnist.data[20000:21000]:
    d = d.reshape(1,1,28,28).astype(np.float32)
    outputs = caffe2.python.onnx.backend.run_model(model, d)
    for x in outputs:
        print(np.argmax(x), end=',')
```
